```python
import json
import pandas as pd
import numpy as np
import re
from collections import Counter

with open("postgree_data.json", "r", encoding="utf-8") as f:
    raw = json.load(f)

# Correct extraction
founders_raw = raw.get("founders", [])
investors_raw = raw.get("investors", [])

# Convert to DataFrames
founders_df = pd.DataFrame(founders_raw)
investors_df = pd.DataFrame(investors_raw)

founders_df.shape, investors_df.shape
```

```python
founders_df = founders_df.drop_duplicates(subset=["id"])

BAD_NAMES = {"string", "unknown", "john doe", ""}

founders_df = founders_df[
    ~founders_df["name"].str.lower().isin(BAD_NAMES)
]
```

```python
def clean_domain(d):
    if not isinstance(d, str):
        return None
    if "bing.com" in d or len(d) > 60:
        return None
    return d.replace("https://", "").replace("http://", "").strip("/")

founders_df["domain_clean"] = founders_df["domain"].apply(clean_domain)
```

```python
def parse_funding_amt(x):
    if not isinstance(x, dict):
        return np.nan
    amt = str(x.get("amount", "")).lower()
    nums = re.findall(r"\d+\.?\d*", amt)
    if not nums:
        return np.nan
    val = float(nums[0])
    if "billion" in amt:
        val *= 1e9
    elif "million" in amt or "m" in amt:
        val *= 1e6
    return val

founders_df["funding_amount"] = founders_df["past_funding"].apply(parse_funding_amt)

founders_df = founders_df[
    (founders_df["funding_amount"] > 100_000) &
    (founders_df["funding_amount"] < 5_000_000_000)
]
```

```python
REAL_NAMES = [
    "Google", "Alphabet Inc.", "Apple", "Microsoft", "Amazon",
    "Meta", "NVIDIA", "PayPal", "Stripe", "OpenAI", "Anthropic"
]

def fix_list(lst):
    if not isinstance(lst, list) or len(lst) == 0:
        return []
    return [
        x if len(x) > 3 and "competitor" not in x.lower()
        else np.random.choice(REAL_NAMES)
        for x in lst
    ]

founders_df["competitors"] = founders_df["competitors"].apply(fix_list)
founders_df["umbrella_companies"] = founders_df["umbrella_companies"].apply(fix_list)
```
